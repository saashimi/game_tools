```python
import random

class Person(object):
    def __init__(self):
        self.surname = None
        self.name = None
        
    def get_height(self):
        feet = self.height // 12
        inches = self.height % 12
        print('Height {0} feet, {1} inches'.format(feet, inches))
        
       
class Male(Person):
    def __init__(self, cat='nml'):
        self.cat = cat 
        self.age = None
        self.height = None
    
    def gen_age(self):
        if self.cat == 'nml':
            age = random.randrange(18, 50)
        elif self.cat == 'yng':
            age = random.randrange(0, 18)
        elif self.cat == 'eld':
            age = random.randrange(50, 85)
        self.age = age
    
    def gen_height(self):
        if self.cat == 'nml' or 'eld':
            height = random.randrange(60, 80)
        else:
            height = random.randrange(48, 72)
        self.height = height
    
class Female(Person):
    def __init__(self, cat='nml'):
        self.cat = cat 
        self.age = None
        self.height = None
    
    def gen_age(self):
        if self.cat == 'nml':
            age = random.randrange(18, 40)
        elif self.cat == 'yng':
            age = random.randrange(0, 18)
        elif self.cat == 'eld':
            age = random.randrange(50, 60)
        self.age = age
    
    def gen_height(self):
        if self.cat == 'nml' or self.cat == 'eld':
            height = random.randrange(58, 72)
        else:
            height = random.randrange(48, 66)
        self.height = height
    
```

```python
surnames = [
    'Abel',
    'Baker',
    'Charlie'
]

names_m = [
    'Mike',
    'Ray',
    'John'
]

names_f = [
    'Ida',
    'Nan',
    'Rey'
]
```

```python
def char_gen(s, grp, skip_age=False):
    random.shuffle(surnames)
    random.shuffle(names_m)
    random.shuffle(names_f)
    
    if s == 'male':
        char = Male(grp)
        char.name = names_m[0]
        char.gen_height()
        if skip_age:
            char.gen_age()
                
    if s == 'female':
        char = Female(grp)
        char.name = names_f[0]
        char.gen_height()
        if skip_age:
             char.gen_age()   
    
    char.surname = surnames[0]
    
    #print('Created character {0} {1}'.format(char.name, char.surname))
    return char
        
```

```python
a = char_gen('male', 'yng')
print(a.age)
a.get_height()
```

```python
# Famu gen
class Famu(object):
    
    def __init__(self):
        self.surname = None
        self.max_kid_age = None
        self.members = {
            'grandpa' : None,
            'grandma' : None,
            'dad'     : None,
            'mom'     : None,
            'kids'    : []
        }
        
    def gen_k_age(self, n):
        ages = []
        for num in range(1, n+1):
            age_gen = random.randrange(8, 22)
            ages.append(age_gen)
        ages.sort()
        return ages

    def gen_parent_age(self, input):
        # Assigns age to lists or int inputs
        if type(input) == list:
            base_age = input[-1] + random.randrange(16, 40)
        elif type(input) == int:
            base_age = input + random.randrange(16, 40)
        return base_age

    def sp_age(self, age):
        calc_age = random.randrange(round(age + age * -.2), round(age + age * .2))
        return calc_age            
        
    def gen_surname(self):
        random.shuffle(surnames)
        self.surname = surnames[0]
        
    def gen_gramps(self):
        self.members['grandpa'] = char_gen('male', 'eld', skip_age=False)
        self.members['grandpa'].age = self.gen_parent_age(self.members['dad'].age)
        self.members['grandma'] = char_gen('female', 'eld', skip_age=False)
        self.members['grandma'].age = self.sp_age(self.members['grandpa'].age)
        
    def gen_parents(self):
        self.members['dad'] = char_gen('male', 'dad', skip_age=False)
        self.members['dad'].age = self.gen_parent_age(self.max_kid_age)
        self.members['mom'] = char_gen('female', 'mom', skip_age=False)
        self.members['mom'].age = self.sp_age(self.members['dad'].age)
    
    def gen_kids(self):
        s_kid = ['male', 'female']
        num_kids = random.randrange(1, 5)
        ages = self.gen_k_age(num_kids)
        for num in range(len(ages)):
            random.shuffle(s_kid)
            k = char_gen(s_kid[0], 'yng', skip_age=True)
            k.age = ages[num]
            self.members['kids'].append(k)
        self.max_kid_age = ages[-1]
            
    def summ_fam_members(self):
        for key, val in self.members.items():
            if key == 'kids':
                for kid in val:
                    print(kid.surname, kid.name, kid.age, kid.get_height())
            else: 
                print(key, val.surname, val.name, val.age, val.height)

```

```python
def gen_famu():
    new_f = Famu()
    new_f.gen_kids()
    new_f.gen_parents()
    new_f.gen_gramps()
    return new_f
```

```python
a = gen_famu()
a.summ_fam_members()
```
